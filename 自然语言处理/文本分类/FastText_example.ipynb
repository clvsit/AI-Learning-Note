```python
import pandas as pd
import numpy as np
import fasttext
```

```python
comment_pos_list, comment_neg_list = [], []

with open('./data/waimai_pos.txt', 'r', encoding='utf-8') as file:
    comment_pos_list = [line.replace('\n', '').strip() for line in file.readlines()]
    
with open('./data/waimai_neg.txt', 'r', encoding='utf-8') as file:
    comment_neg_list = [line.replace('\n', '').strip() for line in file.readlines()]
```

```python
comment_pos_list[:10]
```

```python
comment_neg_list[:10]
```

### 组织成新的训练数据

```python
def label_comment(comment_list, label):
    return [label + comment for comment in comment_list]
```

```python
comment_list = label_comment(comment_pos_list, '__label__1 ')
```

```python
comment_list[:10]
```

```python
comment_list.extend(label_comment(comment_neg_list, '__label__0 '))
```

```python
comment_list[-10:]
```

```python
with open('./data/waimai_train.txt', 'w', encoding='utf-8') as file:
    file.writelines([comment + '\n' for comment in comment_list])
```

### FastText 文本分类

```python
model = fasttext.train_supervised('./data/waimai_train.txt')
```

```python
model.words
```

```python
model.labels
```

```python
model.predict('非常快捷，而且价格也很实惠，好评')
```

### 分词后的尝试

```python
import jieba
```

```python
with open('./dict/stopwords.txt', 'r', encoding='utf-8') as file:
    stopwords = [line.replace('\n', '').strip() for line in file.readlines()]
```

```python
def word_cut(comment, stopwords):
    word_list = []
    
    for word in jieba.lcut(comment):
        if word not in stopwords and word not in ['…']:
            word_list.append(word)
    return word_list
```

```python
new_comment_pos_list = [' '.join(word_cut(comment, stopwords)) for comment in comment_pos_list]
```

```python
new_comment_pos_list[:10]
```

```python
new_comment_neg_list = [' '.join(word_cut(comment, stopwords)) for comment in comment_neg_list]
```

```python
new_comment_neg_list[:10]
```

```python
new_comment_list = label_comment(new_comment_pos_list, '__label__1 ')
```

```python
new_comment_list.extend(label_comment(new_comment_neg_list, '__label__0 '))
```

```python
new_comment_list[:10]
```

```python
new_comment_list[-10:]
```

```python
with open('./data/new_waimai_train.txt', 'w', encoding='utf-8') as file:
    file.writelines([comment + '\n' for comment in new_comment_list])
```

```python
new_model = fasttext.train_supervised('./data/new_waimai_train.txt')
```

```python
new_model.words[:10]
```

```python
new_model.labels
```

```python
new_model.predict(' '.join(word_cut('非常快捷，而且价格也很实惠，好评', stopwords)), k=3)
```
